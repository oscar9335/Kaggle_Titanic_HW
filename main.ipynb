### Age 和 Sex 分析
選取原因:

方法:

  - 先看Age和Sex的缺失值，發現Age有缺失要補值，使用sklearn的module以現有資料學習並補值
- Sex:
  - 使用**轉換類別資料**，將string資料male、female轉成數字後便於分析與訓練。使用Label Encoding。

發現:

````python
import matplotlib
import matplotlib.pyplot as plt
print(matplotlib.__version__)
````

````python
# 資料分析與前處理
# if not using .copy() some warning happen
train_x = df[['Sex', 'Age']].copy()                   # 取出訓練資料需要分析的資料欄位

agedf = df[['Age','Survived']].copy()
sexdf = df[['Sex', 'Survived']].copy()

print("Sex null: {number} ".format(number = sexdf['Sex'].isna().sum()))
print("Age null: {number} ".format(number = agedf['Age'].isna().sum()))

train_y = df['Survived'].copy()                       # 取出訓練資料的答案


from sklearn.impute import SimpleImputer       # 匯入填補缺失值的工具
from sklearn.preprocessing import LabelEncoder # 匯入 Label Encoder

imputer = SimpleImputer(strategy='median')     # 創造 imputer 並設定填補策略
age = train_x['Age'].to_numpy().reshape(-1, 1)   # reshape -1 代表unknow dimension 
imputer.fit(age)                      # 根據資料學習需要填補的值
train_x['Age'] = imputer.transform(age)        # 填補缺失值
agedf['Age'] = imputer.transform(age)

le = LabelEncoder()                            # 創造 Label Encoder
le.fit(train_x['Sex'])                         # 給予每個類別一個數值
train_x['Sex'] = le.transform(train_x['Sex'])  # 轉換所有類別成為數值 (Male,Female) = (1,0)

# le.fit(sexdf['Sex']) 
sexdf['Sex'] = le.transform(sexdf['Sex'])
````

````python
sex_x = [1,0]
male_survived = ((sexdf['Sex'] == 1) & (sexdf['Survived'] == 1))
male_dead = ((sexdf['Sex'] == 1) & (sexdf['Survived'] == 0))

female_survived = ((sexdf['Sex'] == 0) & (sexdf['Survived'] == 1))
female_dead = ((sexdf['Sex'] == 0) & (sexdf['Survived'] == 0))

sex_survived_rate = [male_survived.values.sum()/(male_survived.values.sum() + male_dead.values.sum()),
            female_survived.values.sum()/(female_survived.values.sum() + female_dead.values.sum())]
print(sex_survived_rate)
x = sex_x
y = sex_survived_rate

plt.bar(
    x, 
    y,           
    color='blue',
)

plt.xlabel('Gender ')
plt.ylabel('Survived rate')
plt.xticks(x,['male','female'])  

````

### Pclass 選取與分析
選取原因:
- 上網查找了鐵達尼號wiki，事件發生時負責一等艙的服務人員之需要告知和疏散1等艙乘客即可，而其他服務人員則必須負責告知和疏散2等級3等艙則必須負責告知和疏散2等級3等艙1等艙的的乘客，乘客獲得資訊勢必會較慢，且工作人員會更人手不足。
- 通常較高等的位置享有待遇都會較好。

方法:
- 算出各Pclass的存活率，觀察此feature和存活的相關性。

發現:
- 越好的艙等存活率越高。



````python
pclass_x = [1,2,3]

class1_survive = ((pclass['Pclass'] == 1) & (pclass['Survived'] == 1))
class1_dead = ((pclass['Pclass'] == 1) & (pclass['Survived'] == 0))

class2_survive = ((pclass['Pclass'] == 2) & (pclass['Survived'] == 1))
class2_dead = ((pclass['Pclass'] == 2) & (pclass['Survived'] == 0))

class3_survive = ((pclass['Pclass'] == 3) & (pclass['Survived'] == 1))
class3_dead = ((pclass['Pclass'] == 3) & (pclass['Survived'] == 0))



pclass_y_survive_rate = [class1_survive.values.sum()/(class1_survive.values.sum() + class1_dead.values.sum()),
              class2_survive.values.sum()/(class2_survive.values.sum() + class2_dead.values.sum()),
              class3_survive.values.sum()/(class3_survive.values.sum() + class3_dead.values.sum())] 
print(pclass_y_survive_rate)

x = pclass_x
y = pclass_y_survive_rate

plt.bar(
    x, 
    y,           
    color='blue',
)
plt.xlabel('pclass # ')
plt.ylabel('Survived ratw')
plt.xticks(x,[1,2,3])   
 
````

### Fare 選取與分析
選取原因:
- 票價高通常有較好艙等，有錢人在災難中生存率也較高


方法:
- **轉換數值資料**: 在train裡面發現最高票價和最低票價相差非常多，差500多，將對數值取log
- **轉換類別資料**: 再將其值分成幾的區間觀察(區間數字越大票價越貴)，用於觀察，後面訓練模型時應該不需要分區間

發現:
- 可以看到區間1,2都是死亡大於存活多，區間3生存率來到大概1/3，區間4已經快到0.5，區間5,6都是存活率>0.5了



````python
from sklearn.preprocessing import FunctionTransformer # 匯入轉換數值的工具 - log

transformer = FunctionTransformer(np.log)             # 創造轉換數值工具
fare['Fare'] = transformer.transform(1 + fare['Fare'])     #避免log0
````

## 模型演算法
**Decision Tree**：
 - 有classification trees和regression trees.這邊使用的是前者，利用特徵條件分割出node(survived = 0,survived = 1)，在超參數可以選擇'gini' 或 'entropy'，來決定最佳化決策樹的生成。
 - 在超參數中我使用了
  - criterion='entropy':選這個因為相比於gini有更高準確率。

  - max_depth=10: 試過5、10、15，10結果最好。
  - max_leaf_nodes=32:試過4、8、16、32、64，32結果最好
  - random_state=1012:此notebook原本

**Random Forest**：
- 當成進階版的決策樹，所謂的森林就是由很多棵決策樹所組成。隨機森林是使用 Bagging 加上隨機特徵採樣的方法所產生出來的整體學習演算法。
  1. 從訓練集中抽取 n 筆資料出來
  2. n 筆資料隨機挑選 k 個特徵做樣本
  3. 重複 m 次，產生 m 棵決策樹
  4. Classification: 多數投票機制進行預測、
    Regression: 平均機制進行預測
- 超參數選取:
  - max_depth=10: 試過5、10、15，10結果最好。
  - max_leaf_nodes=32: 試過4、8、16、32、64，32結果最好
  - random_state=1012: 此notebook disision tree原本

**SVC**：
- SVM是一種監督式的學習方法，用統計風險最小化的原則來估計一個分類的超平面(hyperplane)，其基礎的概念非常簡單，就是找到一個決策邊界(decision boundary)讓兩類之間的邊界(margins)最大化，使其可以完美區隔開來。
- 超參數選擇:
  - C=1500:試過500,1000,1500,2000:1500最好
  - random_state=1012: 此notebook disision tree原本

**NearestNeighbors**:
- k是一個用戶定義的常數。一個沒有類別標籤的向量（查詢或測試點）將被歸類為最接近該點的k個樣本點中最頻繁使用的一類。

  簡單來說，我們要找出和新數據附近的K個鄰居（資料），這些鄰居是哪一類（標籤）的它就是哪一類。

````python
# 模型訓練

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.ensemble import RandomForestClassifier   # 匯入ensemble模型
from sklearn.linear_model import LinearRegression    # 匯入regression模型
from sklearn.svm import SVC             # 匯入SVM模型
from sklearn.neighbors import KNeighborsClassifier # 匯入KNeighborsClassifier模型

from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

#tree
train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []
#ensemble
train_ensemble_acc_list = []
valid_ensemble_acc_list = []
# regression
train_regression_acc_list = []
valid_regression_acc_list = []
# svm
train_svm_acc_list = []
valid_svm_acc_list = []

train_neighbors_acc_list = []
valid_neighbors_acc_list = []

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    
    # DecisionTreeClassifier
    model = DecisionTreeClassifier(criterion='entropy',max_depth=10,max_leaf_nodes=32,random_state=1012) # 創造決策樹模型
    model.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = model.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                              train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                              valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

    # ensemble
    ensemble = RandomForestClassifier(max_depth=10,max_leaf_nodes=32,random_state=1012)
    ensemble.fit(train_x_split, train_y_split)

    train_ensemble_pred_y = ensemble.predict(train_x_split)
    train_ensemble_acc = accuracy_score(train_y_split, train_ensemble_pred_y)

    valid_ensemble_pred_y = ensemble.predict(valid_x_split)
    valid_ensemble_acc = accuracy_score(valid_y_split, valid_ensemble_pred_y)

    train_ensemble_acc_list.append(train_ensemble_acc)
    valid_ensemble_acc_list.append(valid_ensemble_acc)


    # svm
    svc_model = SVC(C=1500, random_state=1012)
    svc_model.fit(train_x_split, train_y_split)

    train_svc_pred_y = svc_model.predict(train_x_split)
    train_svc_acc = accuracy_score(train_y_split, train_svc_pred_y)

    valid_svc_pred_y = svc_model.predict(valid_x_split)
    valid_svc_acc = accuracy_score(valid_y_split, valid_svc_pred_y)

    train_svm_acc_list.append(train_svc_acc)
    valid_svm_acc_list.append(valid_svc_acc)

    # NearestNeighbors
    neighbors_model = KNeighborsClassifier()
    neighbors_model.fit(train_x_split, train_y_split)

    train_neighbors_pred_y = neighbors_model.predict(train_x_split)
    train_neighbors_acc = accuracy_score(train_y_split, train_neighbors_pred_y)

    valid_neighbors_pred_y = neighbors_model.predict(valid_x_split)
    valid_neighbors_acc = accuracy_score(valid_y_split, valid_neighbors_pred_y)

    train_neighbors_acc_list.append(train_neighbors_acc)
    valid_neighbors_acc_list.append(valid_neighbors_acc)



print("DecisionTreeClassifier")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
print()
print("RandomForestClassifier(ensemble)")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_ensemble_acc_list),                          # 輸出平均訓練準確度
    np.min(train_ensemble_acc_list),                           # 輸出最低訓練準確度
    np.max(train_ensemble_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_ensemble_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_ensemble_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_ensemble_acc_list)                            # 輸出最高驗證準確度
))
print()
print("SVM")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_svm_acc_list),                          # 輸出平均訓練準確度
    np.min(train_svm_acc_list),                           # 輸出最低訓練準確度
    np.max(train_svm_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_svm_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_svm_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_svm_acc_list)                            # 輸出最高驗證準確度
))
print()
print("NearestNeighbors")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_neighbors_acc_list),                          # 輸出平均訓練準確度
    np.min(train_neighbors_acc_list),                           # 輸出最低訓練準確度
    np.max(train_neighbors_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_neighbors_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_neighbors_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_neighbors_acc_list)                            # 輸出最高驗證準確度
))
````

## Overfitting(DecisionTree 方法)

DecisionTreeClassifier中 
- min_samples_split:
 - 如果沒設定就是2，也就是說使得資料數最少要只要2比就可以 split 新的節點。
- min_samples_leaf:
 - 任何一個葉節點的資料數不得小於幾個leaf幾個leaf，預設為1，會產生很多節點，容易overfitting。



````python
from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型

from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

#tree
train_acc_overfitting_list = []    # 儲存每次訓練模型的準確度
valid_acc_overfitting_list = []

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    
    # DecisionTreeClassifier
    overmodel = DecisionTreeClassifier(random_state=1012) # 創造決策樹模型
    overmodel.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_predover_y = overmodel.predict(train_x_split)       # 確認模型是否訓練成功
    train_overacc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                              train_predover_y)
    valid_predover_y = overmodel.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_overacc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                              valid_predover_y)
    
    train_acc_overfitting_list.append(train_overacc)
    valid_acc_overfitting_list.append(valid_overacc)

print("DecisionTreeClassifier Overfitting")
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_overfitting_list),                          # 輸出平均訓練準確度
    np.min(train_acc_overfitting_list),                           # 輸出最低訓練準確度
    np.max(train_acc_overfitting_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_overfitting_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_overfitting_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_overfitting_list)                            # 輸出最高驗證準確度
))
print()
````
